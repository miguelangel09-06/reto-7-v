4.Imprimir los números de 1 hasta un número natural n dado, cada uno con su respectivo factorial

```python
#Primero se crea una funcion que va a determinar los factoriales 

def facto(ingresado):
    contador=1 #Esta variable va a llevar la cuenta de la multiplicacion, es decir, en ella recae el 1 x 2 x 3 x 4 ....
    for i in range(1,ingresado+1):#Se crea un rango el cual es del tamaño del numero a evaluar, es decir, si ingresa el 4, el rango va a ser de 1 a 4
        contador*=i #Aqui se multiplican los numeros, empieza 1 x 2 y luego 2 x 3, despues 6 x 4.. asi se va cumpliendo el factorial, por ejemplo si es 4, su factorial es 24
    return contador


if __name__ == "__main__":
 


 ingresado=int(input("Por favor ingrese un numero="))# se pide el numero
if ingresado<2:
    print("Numero no valido")#Acaba el programa si el numero no es apto
      
for i in range(1,ingresado+1): #este for recorre la lista de los numeros, de 1 al natural determinado,lee numero por numero a su vez que imrpime el numero y a cada numero le ejecuta la funcion del principio.
    print(i,facto(i))
```

5.Calcular el valor de 2 elevado a la potencia n usando ciclos for.

```python
#Este codigo es a base del punto anterior 

pedido=int(input("Por favor ingrese un numero el cual va a ser el exponente de 2="))#Se pide el numero del exponente
if pedido<0:
    print("No valido para este caso :)")#Acaba si no es apto


exp=1  #Esta variable lleva la cuenta de la multiplicacion
for i in range(pedido):#Se establece un rango hasta el numero que se ingresa 
    exp*=2 #Se multiplica el 2 por 1 al empezar, entonces valdria 2 , luego vuelve a ingresar y se vuelve a multiplicar por 2 las veces que el numero del rango establezca, osea que si ingreso un  2 el resultado es 4 (1 x 2= 2, 2 x 2 =4)

print("2 ^",pedido,"=",exp)#Se imprime el resultado
```

8.Diseñar una función que permita calcular una aproximación de la función exponencial alrededor de 0 para cualquier valor x (real), utilizando los primeros n términos de la serie de Taylor. Nota: use math para traer la función exponencial y mostrar la diferencia entre el valor real y la aproximación. 

```python
import math #Importo la libreria para hacer las operaciones matematicas 

def calcular_aproximacion(x,n): #Se crea esta funcion que va a calcular basicamente la aproximacion, dentro de esta funcion practicamente va la formula puesta en el repo

    cont=0 # Coloco una varibale que va a llevar la cuenta de la suma 
    for i in range(0,n+1): #Este for recorre desde i=0 hasta n (En otras palabras la sumatoria que esta en la formula)
        cont1=x**i/ math.factorial(i) #Esta operacion es la de la formula para calcular la aproximacion, se importa la funcion para calcular el factorial
        cont+=cont1 #Aqui se lleva la sumatoria 
    return cont #El resultado va aqui

if __name__ == "__main__":  
#Se piden los datos
    x=float(input("Por favor ingrese la variable X"))
n=int(input("Por favor ingrese la cantidad de terminos relacionados con la exactitud de la aproximacion="))

provicional:float=calcular_aproximacion(x,n) #Esta variable es para estar mas claros
valor_real = math.exp(x) #El valor real de la funcion exponencial se calcula mediante esta funcion (es e ^ x )
diferencia_exp= valor_real-provicional #Se restan el valor real y las aproximacion


#Se imprimen los resultados
print("Aproximacion de e ^",x, "con ",n, " terminos es= ", calcular_aproximacion(x,n))
print("El valor real es = ", valor_real)
print("La diferencia entre el valor real y la aproximacion es = ",diferencia_exp)
```
